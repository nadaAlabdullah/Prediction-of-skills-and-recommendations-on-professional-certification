```python
# main calulcation funtion
def calculate(skills_list, export_prefix):
    df = pd.DataFrame({'skills_list': skills_list})
    
    # counts
    df = df.groupby(['skills_list']).size().reset_index(name='count').copy()
    
    # max and min
    print('{} skills'.format(export_prefix))
    print('Max: {}'.format(df['count'].max()))
    print('Min: {}'.format(df['count'].min()), end='\n\n')
    
    # normalise (rate)
    df['rate'] = (df['count']-df['count'].min())/(df['count'].max()-df['count'].min())*5
    df
    
    # export normalise results to csv
    filename = export_prefix + '_Norm_results.csv'
    df.to_csv('files/'+filename)
    
    # categorise (Label Encoder)
    le = LabelEncoder()
    df['skills_list_catogrised']=le.fit_transform(df['skills_list'])
    df
    
    # export final results to csv
    filename = export_prefix + '_final.csv'
    df.to_csv('files/'+filename)
    
    # export final results to xlsx
    filename = export_prefix + '_final.xlsx'
    df.to_excel('files/'+filename)
```
